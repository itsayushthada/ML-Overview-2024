## Cummulative Sum

```python
# Cummulative Sum over single dataframe column
df[['Revenue']].cumsum(skipna=True)
```

```python
# Cummulative Sum over multiple dataframe column
df[['Units_Sold', 'Revenue', 'Discount_Percentage']].cumsum(skipna=True)
```

```python
# Cummulative Sum within groups
df.groupby(['Product_Type', 'Region'])[['Units_Sold', 'Revenue', 'Discount_Percentage']].apply(lambda x: x.cumsum()).sort_index()
```

## Window Functions

#### Rolling Window
Rolling window: Generic fixed or variable sliding window over the values.



```python
# Rolling mean. Min_periods as window will not be full always
df['Revenue'].rolling(3, min_periods=1).mean()
```

```python
# manually veruying mean with averages
df['Revenue'].rolling(3, min_periods=1).sum() / df['Revenue'].rolling(3, min_periods=1).count() == df['Revenue'].rolling(3, min_periods=1).mean()
```

```python
# Window is ignoring nulls
df['Discount_Percentage'].rolling(3, min_periods=1).count()
```

```python
# Running T Score

df['Rolling_Mean_Discount_Percentage'] = df['Discount_Percentage'].rolling(3, min_periods=1).apply(lambda x: x.mean())
df['Rolling_Std_Discount_Percentage'] = df['Discount_Percentage'].rolling(3, min_periods=1).apply(lambda x: x.std())
df['Rolling_count_Discount_Percentage'] = df['Discount_Percentage'].rolling(3, min_periods=1).count()

df['Rolling_count_Discount_Percentage']*(df['Discount_Percentage'] - df['Rolling_Mean_Discount_Percentage'])/df['Rolling_Std_Discount_Percentage']
```

#### Expanding Window

: This creates an expanding window where the number of observations grows with each step.

```python
# Running T Score

df['Rolling_Mean_Discount_Percentage'] = df['Discount_Percentage'].expanding(min_periods=1).apply(lambda x: x.mean())
df['Rolling_Std_Discount_Percentage'] = df['Discount_Percentage'].expanding(min_periods=1).apply(lambda x: x.std())
df['Rolling_count_Discount_Percentage'] = df['Discount_Percentage'].expanding(min_periods=1).count()

df['Rolling_count_Discount_Percentage']*(df['Discount_Percentage'] - df['Rolling_Mean_Discount_Percentage'])/df['Rolling_Std_Discount_Percentage']
```

```python
# We ca see mean and standard deviatin getting stable as the window size grows.
df[['Discount_Percentage', 'Rolling_Mean_Discount_Percentage', 'Rolling_Std_Discount_Percentage', 'Rolling_count_Discount_Percentage']]
```
