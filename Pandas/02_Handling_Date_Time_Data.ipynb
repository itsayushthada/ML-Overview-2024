#### Interpolation

Convert date tiem into timestamp and then interpolate. Convert timestamp back to datetime

```python
df = get_df()
df
```

```python
def interpolate(column):
  df['TS_col'] = column.astype('int64')
  df['TS_col'] = df['TS_col'].apply(lambda x: None if x<0 else x)
  df['TS_col'] = df['TS_col'].interpolate(method='linear')
  column = df['TS_col'].astype('datetime64[ns]')
  df.drop('TS_col', axis=1, inplace=True)
  return column
```

```python
for x in formats:
  df[x] = interpolate(df[x])
```

#### Average of last and next available Values

To calculate the average of two dates (i.e., the midpoint), you cannot directly use mean() on a datetime64 column. Instead, you need to convert the datetime64 values to an integer representation (e.g., nanoseconds), compute the mean, and then convert it back to a datetime.

```python
df = get_df()
for key in formats:
  df[key] = pd.to_datetime(df[key], errors='coerce')
```

```python
df['ffill_ES'] = df['Experiment_Start'].fillna(method='ffill')
df['bfill_ES'] = df['Experiment_Start'].fillna(method='bfill')
```

```python
df['ES_New'] = ((df['ffill_ES'].astype('int64') + df['bfill_ES'].astype('int64'))//2).astype('datetime64[ns]')
idx = df['ffill_ES'].isna()
df.loc[idx, 'ES_New'] = df.loc[idx, 'bfill_ES']

idx = df['bfill_ES'].isna()
df.loc[idx, 'ES_New'] = df.loc[idx, 'ffill_ES']
```
