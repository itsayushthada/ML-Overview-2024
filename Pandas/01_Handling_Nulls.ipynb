## Impute With Simple Imputer [Single Imputation]

```python
df = get_dataframe()
```

```python
# Mean Imputer
imp_mean = impute.SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(df)
pd.DataFrame(imp_mean.transform(df), columns=df.columns)
```

```python
# Median Imputer
imp_mean = impute.SimpleImputer(missing_values=np.nan, strategy='median')
imp_mean.fit(df)
pd.DataFrame(imp_mean.transform(df), columns=df.columns)
```

```python
# Constant Imputer
imp_mean = impute.SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=-1)
imp_mean.fit(df)
pd.DataFrame(imp_mean.transform(df), columns=df.columns)
```

```python
# Most Frequent Imputer
imp_mean = impute.SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mean.fit(df)
pd.DataFrame(imp_mean.transform(df), columns=df.columns)
```

## Impute With MICE [Multivaraite Imputation by Chain Method]

```python
df = get_dataframe()
```

```python
# IterativeImputer Imputer
imp_mice = impute.IterativeImputer(max_iter=10, random_state=0, initial_strategy='mean')
imp_mice.fit(df)
pd.DataFrame(imp_mice.transform(df), columns=df.columns)
```

```python
# IterativeImputer Imputer
imp_mice = impute.IterativeImputer(max_iter=3, random_state=0, initial_strategy='mean')
imp_mice.fit(df)
pd.DataFrame(imp_mice.transform(df), columns=df.columns)
```

```python
# IterativeImputer Imputer
imp_mice = impute.IterativeImputer(max_iter=10, random_state=0, initial_strategy='most_frequent')
imp_mice.fit(df)
pd.DataFrame(imp_mice.transform(df), columns=df.columns)
```

## Imputer with KNN [Best Imputation method*][Multiple Imputer]

```python
df = get_dataframe()
```

```python
# KNN Imputer
imp_knn = impute.KNNImputer(n_neighbors=3, weights="uniform")
imp_knn.fit(df)
pd.DataFrame(imp_knn.transform(df), columns=df.columns)
```

```python
# KNN Imputer
imp_knn = impute.KNNImputer(n_neighbors=5, weights="uniform")
imp_knn.fit(df)
pd.DataFrame(imp_knn.transform(df), columns=df.columns)
```

```python
# KNN Imputer
imp_knn = impute.KNNImputer(n_neighbors=3, weights="distance")
imp_knn.fit(df)
pd.DataFrame(imp_knn.transform(df), columns=df.columns)
```
